```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='r',
         linestyle='--', label='Tüm Yazılımcılar')

plt.plot(ages, py_salaries, label='Python')

plt.legend()
plt.xlabel('Yaşlar')
plt.ylabel('Maaşlar')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='r',
         linestyle='--', label='Tüm Yazılımcılar')

plt.plot(ages, py_salaries, label='Python')

plt.fill_between(ages, py_salaries, alpha=0.25)

plt.legend()
plt.xlabel('Yaşlar')
plt.ylabel('Maaşlar')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='r',
         linestyle='--', label='Tüm Yazılımcılar')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25)

plt.legend()
plt.xlabel('Yaşlar')
plt.ylabel('Maaşlar')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='r',
         linestyle='--', label='Tüm Yazılımcılar')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, alpha=0.25)
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries <= overall_median), interpolate=True,alpha=0.25)

plt.legend()
plt.xlabel('Yaşlar')
plt.ylabel('Maaşlar')
plt.tight_layout()
plt.show()
```
