```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')

data = pd.read_csv('data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, ax = plt.subplots()

ax.plot(ages, dev_salaries, color='r',
         linestyle='--', label='Tm Yazlmclar')

ax.plot(ages, py_salaries, color="b", label='Python')
ax.plot(ages, js_salaries, color="y", label='JavaScript')

ax.legend()

ax.set_xlabel("Yalar")
ax.set_ylabel("Maalar")

plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

data = pd.read_csv('data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(ages, dev_salaries, color='r',
         linestyle='--', label='Tm Yazlmclar')

ax2.plot(ages, py_salaries, color="b", label='Python')
ax2.plot(ages, js_salaries, color="y", label='JavaScript')

ax1.legend()
ax1.set_ylabel("Maalar")

ax2.legend()
ax2.set_xlabel("Yalar")
ax2.set_ylabel("Maalar")

plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

data = pd.read_csv('data2.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='r',
         linestyle='--', label='Tm Yazlmclar')

ax2.plot(ages, py_salaries, color="b", label='Python')
ax2.plot(ages, js_salaries, color="y", label='JavaScript')

ax1.legend()
ax1.set_ylabel("Maalar")

ax2.legend()
ax2.set_xlabel("Yalar")
ax2.set_ylabel("Maalar")

plt.tight_layout()
plt.show()

fig1.savefig("fig1.png")
fig2.savefig("fig2.png")
```
