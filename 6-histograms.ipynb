```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.hist(ages, bins=5, edgecolor="black")

plt.xlabel('Yaşlar')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50, 60]

plt.hist(ages, bins=bins, edgecolor="black")

plt.xlabel('Yaşlar')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50, 60]
median_age = 29

plt.hist(ages, bins=bins, edgecolor="black")

plt.axvline(median_age, color="r", label="Orta Yaş")

plt.legend()
plt.xlabel('Yaşlar')
plt.tight_layout()
plt.show()
```
